# Plota estratégia sem terreno

```python
# cria moldura da figura
fig = plt.figure(figsize=(10,8))

# define estilo do gráfico
plt.style.use('bmh')

#========================================================================================#
#                                    BANDA-S
#========================================================================================#
#--------------------------------------------#
#              Parâmetros do radr
#--------------------------------------------#
# latitude e longitude do radar [graus]
lon_radar, lat_radar = -45.5825, -22.5344

# altitude do radar [metros]
alt_radar = 1850

# ângulos de elevação [graus]
sweeps = [1.0, 1.6, 2.4, 3.2, 4.2, 5.5, 6.9, 8.6]
sweeps = [-0.5, 0.0, 0.5, 1.0, 1.6, 2.4, 3.2, 4.2, 5.5, 6.9, 8.6]

# quantidade de bins do feixe [quantidade]
ngates = 960

# resolução radial [metros]
radial_resolution = 250.0

# distância de cada bin ao radar [metros]
ranges = np.arange(ngates) * radial_resolution

# ângulo de abertura do feixe [graus]
radar_beam_width = 0.968

# coordenadas do radar (longitude, latitude, altitude) [graus, graus, metros]
sitecoords = (lon_radar, lat_radar, alt_radar)

#--------------------------------------------#
#              Plota figura
#--------------------------------------------#
# moldura da figura
ax1 = plt.subplot(211)

# plota figura
wrl.vis.plot_scan_strategy(ranges,
                           sweeps,
                           sitecoords,
                           beamwidth=radar_beam_width,
                           vert_res = 1000.,
                           maxalt = 15000.,
                           range_res = 10000.,
                           maxrange = 240000.,
                           units='km',
                           ax=ax1)

# plota linha cortando em 100km
#ax1.axvline(x=100000., color='gray', linestyle='-', linewidth=2)
#ax1.text(x=100000., y=2500., s='100km', color='gray')

# plota linha cortando em ITAJUBÁ
ax1.axvline(x=20000., color='red', linestyle='-', linewidth=2)
ax1.text(x=21000., y=2300., s='Itajubá', color='red')

# plota linha cortando em 2 e 3km de altitude
ax1.axhline(y=2000., color='red', linestyle='--', linewidth=2)
ax1.axhline(y=3000., color='red', linestyle='--', linewidth=2)

# legenda
leg1 = f'ngates = {ngates}'
leg2 = f'radial_resolution = {radial_resolution} m'
leg3 = f'radar_beam_width = {radar_beam_width} $\degree$'
ax1.text(6000,
         12500.,
         f'Configuração do Radar\n{leg1}\n{leg2}\n{leg3}',
         size=7,
         bbox=dict(facecolor='lightgray', edgecolor='black', boxstyle='round,pad=0.5'))

# título
ax1.set_title('Banda-S')

# nomes dos eixos X e Y
ax1.set_xlabel(' ')
ax1.set_ylabel('Altura (km)')

# mantém apenas os marcadores a cada 5 posições, reduzindo a quantidade de marcas exibidas no eixo X
ax1.set_xticks(ax1.get_xticks()[::5])

# limites dos eixos X e Y
ax1.set_xlim(0., 240000.)
ax1.set_ylim(0., 15000.)

# grade da figura
ax1.grid()

# não plota os "minorticks" (menores marcadores)
ax1.minorticks_off()

#========================================================================================#
#                                BANDA-X
#========================================================================================#
#--------------------------------------------#
#              Parâmetros do radr
#--------------------------------------------#
# latitude e longitude do radar [graus]
lon_radar, lat_radar = -45.5825, -22.5344

# altitude do radar [metros]
alt_radar = 1850

# ângulos de elevação [graus]
sweeps = [1.0, 1.7, 2.6, 3.6, 4.8, 6.2, 7.8, 9.7, 11.9, 14.5, 17.5, 20.9, 25.0]
sweeps = [-0.5, 0.0, 0.5, 1.0, 1.7, 2.6, 3.6, 4.8, 6.2, 7.8, 9.7, 11.9, 14.5, 17.5, 20.9, 25.0]
sweeps = [-0.5, 0.0, 0.5, 1.0, 1.6, 2.4, 3.2, 4.2, 5.5, 6.9, 8.6]

# quantidade de bins do feixe [quantidade]
ngates = 666

# resolução radial [metros]
radial_resolution = 150.0

# distância de cada bin ao radar [metros]
ranges = np.arange(ngates) * radial_resolution

# ângulo de abertura do feixe [graus]
radar_beam_width = 1.3

# coordenadas do radar (longitude, latitude, altitude) [graus, graus, metros]
sitecoords = (lon_radar, lat_radar, alt_radar)

#--------------------------------------------#
#              Plota figura
#--------------------------------------------#
# moldura da figura
ax2 = plt.subplot(212)

# plota figura
wrl.vis.plot_scan_strategy(ranges,
                           sweeps,
                           sitecoords,
                           beamwidth=radar_beam_width,
                           vert_res = 1000.,
                           maxalt = 15000.,
                           range_res = 10000.,
                           maxrange = 100000.,
                           units='km',
                           ax=ax2)

# plota linha cortando em 2km de altitude
ax2.axhline(y=2000., color='red', linestyle='--', linewidth=2)
ax2.axhline(y=3000., color='red', linestyle='--', linewidth=2)

# plota linha cortando em ITAJUBÁ
ax2.axvline(x=20000., color='red', linestyle='-', linewidth=2)
ax2.text(x=21000., y=2300., s='Itajubá', color='red')

# legenda
leg1 = f'ngates = {ngates}'
leg2 = f'radial_resolution = {radial_resolution} m'
leg3 = f'radar_beam_width = {radar_beam_width} $\degree$'
ax2.text(3000,
         12500.,
         f'Configuração do Radar\n{leg1}\n{leg2}\n{leg3}',
         size=7,
         bbox=dict(facecolor='lightgray', edgecolor='black', boxstyle='round,pad=0.5'))

# título
ax2.set_title('Banda-X')

# nomes dos eixos X e Y
ax2.set_xlabel('Distância (km)')
ax2.set_ylabel('Altura (km)')

# mantém apenas os marcadores a cada 2 posições, reduzindo a quantidade de marcas exibidas no eixo X
ax2.set_xticks(ax2.get_xticks()[::2])

# limites dos eixos X e Y
ax2.set_xlim(0., 100000.)
ax2.set_ylim(0., 15000.)

# grade da figura
ax2.grid()

# não plota os "minorticks" (menores marcadores)
ax2.minorticks_off()

# recorta figura
plt.tight_layout()

# salva figura
plt.savefig(f'{dir}/output/codigo_2/estrategia_varredura_radarSeX_v2.jpg',
            dpi=300,
            bbox_inches='tight')
```

# Plota estratégia com terreno - `TESTANDO`
- https://docs.wradlib.org/en/latest/notebooks/visualisation/plot_scan_strategy.html

```python
nrays = 360
nbins = 150
range_res = 500.0
ranges = np.arange(nbins) * range_res
elevs = [28.0, 18.0, 14.0, 11.0, 8.2, 6.0, 4.5, 3.1, 2.0, 1.0]
site = (-28.1, 38.42, 450.0)
beamwidth = 1.0

# only run if environment variables are set
import os

has_data = os.environ.get("WRADLIB_EARTHDATA_BEARER_TOKEN", False)

if has_data:
    ax = wrl.vis.plot_scan_strategy(
        ranges, elevs, site, units="km", terrain=True, az=10
    )
```

```python
from wradlib.io.radolan import wradlib_data # Import wradlib_data
flist = [
    "geo/N38W028.SRTMGL3.hgt.zip",
    "geo/N38W029.SRTMGL3.hgt.zip",
    "geo/N39W028.SRTMGL3.hgt.zip",
    "geo/N39W029.SRTMGL3.hgt.zip",
]
[wradlib_data.DATASETS.fetch(f) for f in flist]
xyz, rad = wrl.georef.spherical_to_xyz(ranges, [10.0], elevs, site, squeeze=True)
ll = wrl.georef.reproject(xyz, src_crs=rad)

# (down-)load srtm data
ds = wrl.io.get_srtm(
    [ll[..., 0].min(), ll[..., 0].max(), ll[..., 1].min(), ll[..., 1].max()],
)
rastervalues, rastercoords, proj = wrl.georef.extract_raster_dataset(
    ds, nodata=-32768.0
)
# map rastervalues to polar grid points
terrain = wrl.ipol.cart_to_irregular_spline(
    rastercoords, rastervalues, ll[-1, ..., :2], order=3, prefilter=False)

ax = wrl.vis.plot_scan_strategy(ranges, elevs, site, units="km", terrain=terrain)
```
